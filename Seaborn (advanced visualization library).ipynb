
# Jointplot 

joint plot allows to study the relationship between two numerical variables.This chart display their correlation.it is usually a scatterplot,a hexbin plot,a 2D Histogram or a density plot

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

# Pairplot

pairplot is also known as 'scatterplot'.it shows all variables paired with all other variables.

```python
sns.pairplot(data=df)
```

```python
sns.pairplot(data=df,hue='sex')
```

# Distplot

distplot helps us to see the distribution of the column feature

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

# categorical plots

seaborn also helps us to analyse catergorical variables

1.boxplot

2.violinplot

3.countplot

4.bar plot

# countplot

```python
sns.countplot(data=df,x='sex')
```

# barplot

```python
sns.barplot(data=df,x='sex',y='total_bill')
```

# box plot

a boxplot represents information from a five-number summary 

```python
sns.boxplot('smoker','total_bill',data=df)
```

# violin plot

violin plot helps us to see the distribution of data in terms of kernel density estimation and the box plot 

```python
sns.violinplot('total_bill','day',data=df)
```
